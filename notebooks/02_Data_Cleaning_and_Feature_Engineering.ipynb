Preparing the dataset for Decision Tree Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style("whitegrid")

print("Libraries imported successfully!")
```

```python
# Load the original dataset
df = pd.read_csv('ds_salaries.csv')
print('Original shape:', df.shape)
df.head()
```

```python
# 1. Drop redundant columns
df_clean = df.drop(columns=['Unnamed: 0', 'salary', 'salary_currency']).copy()
print('Columns after dropping redundant ones:')
print(df_clean.columns.tolist())
```

```python
# 2. Feature Engineering - Job Title Grouping (Very Important!)
def group_job_title(title):
    title = str(title).lower()
    if 'data scientist' in title:
        return 'Data Scientist'
    elif 'data engineer' in title or 'big data' in title or 'etl' in title:
        return 'Data Engineer'
    elif 'data analyst' in title or 'bi data' in title or 'business data' in title or 'analytics' in title:
        return 'Data Analyst'
    elif any(x in title for x in ['machine learning', 'ml engineer', 'mle']):
        return 'ML Engineer'
    elif any(x in title for x in ['director', 'head', 'principal', 'lead']):
        return 'Lead/Director'
    else:
        return 'Other'

df_clean['job_category'] = df_clean['job_title'].apply(group_job_title)

print('Job categories created:')
print(df_clean['job_category'].value_counts())
```

```python
# 3. Create useful binary flags
df_clean['is_us_company'] = (df_clean['company_location'] == 'US').astype(int)
df_clean['is_us_residence'] = (df_clean['employee_residence'] == 'US').astype(int)

print('US Company flag created. US companies:', df_clean['is_us_company'].sum())
```

```python
# 4. Ordinal Encoding for ordered categorical features
exp_order = {'EN': 0, 'MI': 1, 'SE': 2, 'EX': 3}
size_order = {'S': 0, 'M': 1, 'L': 2}

df_clean['experience_level_enc'] = df_clean['experience_level'].map(exp_order)
df_clean['company_size_enc'] = df_clean['company_size'].map(size_order)

print('Ordinal encoding done.')
```

```python
# 5. Drop original high-cardinality columns we replaced
columns_to_drop = ['job_title', 'employee_residence', 'company_location']
df_clean = df_clean.drop(columns=columns_to_drop)

print('Final columns:')
print(df_clean.columns.tolist())
print('\nFinal shape:', df_clean.shape)
```

```python
# 6. Quick final check
print(df_clean.head())
print('\nAny missing values?')
print(df_clean.isnull().sum())
```

```python
# 7. Save the cleaned dataset
df_clean.to_csv('ds_salaries_cleaned.csv', index=False)
print('✅ Cleaned dataset saved as ds_salaries_cleaned.csv')
```

```python
# Optional: Quick visualization of new features
plt.figure(figsize=(10, 6))
sns.boxplot(x='job_category', y='salary_in_usd', data=df_clean)
plt.xticks(rotation=45)
plt.title('Salary Distribution by Job Category (After Grouping)')
plt.show()
```

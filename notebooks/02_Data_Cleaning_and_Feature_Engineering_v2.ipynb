# 02_Data_Cleaning_and_Feature_Engineering (Improved Job Grouping)

```python
import pandas as pd

df = pd.read_csv('ds_salaries.csv')

# Improved Job Title Grouping
def group_job_title(title):
    title = str(title).lower().strip()
    
    if 'data scientist' in title:
        return 'Data Scientist'
    elif 'data engineer' in title or 'big data engineer' in title:
        return 'Data Engineer'
    elif 'data analyst' in title or 'bi data analyst' in title or 'business data analyst' in title:
        return 'Data Analyst'
    elif 'machine learning engineer' in title or 'ml engineer' in title:
        return 'ML Engineer'
    elif 'ai scientist' in title:
        return 'AI Scientist'
    elif 'research scientist' in title:
        return 'Research Scientist'
    elif 'data architect' in title:
        return 'Data Architect'
    elif any(word in title for word in ['director', 'head', 'principal', 'lead', 'manager']):
        return 'Lead/Director'
    else:
        return 'Other'

df_clean = df.drop(columns=['Unnamed: 0', 'salary', 'salary_currency']).copy()

df_clean['job_category'] = df_clean['job_title'].apply(group_job_title)

# Create flags
df_clean['is_us_company'] = (df_clean['company_location'] == 'US').astype(int)
df_clean['is_us_residence'] = (df_clean['employee_residence'] == 'US').astype(int)

# Ordinal encoding
exp_order = {'EN': 0, 'MI': 1, 'SE': 2, 'EX': 3}
size_order = {'S': 0, 'M': 1, 'L': 2}

df_clean['experience_level_enc'] = df_clean['experience_level'].map(exp_order)
df_clean['company_size_enc'] = df_clean['company_size'].map(size_order)

# Drop high-cardinality original columns
df_clean = df_clean.drop(columns=['job_title', 'employee_residence', 'company_location'])

print('New job_category distribution:')
print(df_clean['job_category'].value_counts())

df_clean.to_csv('ds_salaries_cleaned_v2.csv', index=False)
print("\n✅ Saved as ds_salaries_cleaned_v2.csv")
```
